```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
titanic =pd.read_csv("Titanic.csv")
titanic
```

```python
titanic.info()
```

#### Observations
* There are no null values
* All columns are object data type and categorical in nature
* As the columns are categorical, we can do adopt one-hot-encoding

```python
counts=titanic['Class'].value_counts()
plt.bar(counts.index, counts.values)
```

```python
df=pd.get_dummies(titanic,dtype=int)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
frequent_itemsets=apriori(df,min_support=0.05,use_colnames=True,max_len=None)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1.0)
rules
```

```python
rules.sort_values(by='lift',ascending=False)
```

```python
import matplotlib.pyplot as plt
rules[['support','confidence','lift']].hist(figsize=(15,7))
plt.show()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel("Confidence")
plt.show()
```

```python
plt.scatter(rules['confidence'],rules['lift'])
plt.xlabel("Confidence")
plt.ylabel("Lift")
plt.show()
```

```python
plt.scatter(rules['support'],rules['lift'])
plt.xlabel("Support")
plt.ylabel("Lift")
plt.show()
```
